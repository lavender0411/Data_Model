### Analysis
* 根據消費者是否符合信貸標準，其信用評分分佈
* 根據消費者是否全額付款，其信用評分分佈

```python
sns.set_style('whitegrid')
plt.figure(figsize = (12, 6))
df[df['credit.policy']==1]['fico'].hist(
    bins = 35, color = 'blue', 
    label = 'Credit Policy = 1', alpha = 0.6)
df[df['credit.policy']==0]['fico'].hist(
    bins = 35, color = 'red', 
    label = 'Credit Policy = 0', alpha = 0.6)
plt.legend()
plt.xlabel('FICO')
plt.show()
```

```python
plt.figure(figsize = (12, 6))
df[df['not.fully.paid']==1]['fico'].hist(
    bins = 35, color = 'blue', 
    label = 'Not Fully Paid = 1', alpha = 0.6)
df[df['not.fully.paid']==0]['fico'].hist(
    bins = 35, color = 'red', 
    label = 'Not Fully Paid = 0', alpha = 0.6)
plt.legend()
plt.xlabel('FICO')
plt.show()
```

* 根據消費者是否全額付款及貸款目的之數量分佈

```python
plt.figure(figsize = (11, 7))
sns.countplot(x = 'purpose', 
              data = df, hue = 'not.fully.paid', 
              palette = 'Set1')
plt.tight_layout()
plt.show()
```

* 信用評分及一年的利息與本金之間百分比(利率)關係

```python
sns.jointplot(x = 'fico', y = 'int.rate', 
              data = df, color = 'purple')
plt.show()
```

* 信用評分及一年的利息與本金之間百分比(利率)關係，及是否全額付款與是否符合信貸標準

```python
plt.figure(figsize = (11, 7))
sns.lmplot(x = 'fico', y = 'int.rate', 
           data = df, 
           col="not.fully.paid", 
           hue = 'credit.policy')
plt.show()
```

### Model (decision tree)

```python
final = pd.get_dummies(df, columns = ['purpose'], 
               drop_first = True, dtype = int)
final.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final.drop('not.fully.paid', axis = 1)
y = df['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

### Prediction

```python
pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```

### Model (random forest)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=300)
rfc.fit(X_train, y_train)
```

### Predictions

```python
pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```
