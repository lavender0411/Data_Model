# college_data
Dataset from kaggle.com

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
col_data = pd.read_csv('College_Data', 
                       index_col = 0)
col_data.head()
```

```python
print(col_data.info())
col_data.describe()
```

## Cleaning

```python
col_data['Private'] = col_data['Private'].apply(lambda priv: 1 if priv=='Yes' else 0)
```

## Visualization

#### Room.Board cost v.s. Grad.Rate

```python
sns.set_style('whitegrid')

sns.lmplot(x = 'Room.Board', y = 'Grad.Rate', 
                data = col_data, hue = 'Private', 
           fit_reg = False, 
           palette = 'coolwarm', 
           height = 6, aspect = 1
          )
plt.show()
```

#### F.Undergrad v.s. Outstate

```python
sns.lmplot(x = 'Outstate', y = 'F.Undergrad',
           data = col_data, hue = 'Private', 
           fit_reg = False, 
           palette = 'coolwarm', 
           height = 6, aspect = 1)
plt.show()
```

#### Outstate v.s. Private

```python
g = sns.FacetGrid(data = col_data, 
              hue = 'Private', palette = 'coolwarm', 
                  height = 6, aspect = 2, hue_order=[1, 0])
g.map(plt.hist, "Outstate", bins = 20, alpha = 0.7, edgecolor = 'black').add_legend()
plt.show()
```

#### Grad.Rate v.s. Private

```python
g = sns.FacetGrid(data = col_data, 
              hue = 'Private', palette = 'coolwarm', 
                  height = 6, aspect = 2, hue_order=[1, 0])
g.map(plt.hist, "Grad.Rate", bins = 20, alpha = 0.7, 
      edgecolor = 'black').add_legend()
plt.show()
```

#### Private with grad.rate > 100%

```python
col_data[(col_data['Private']==1) & (col_data['Grad.Rate']>100)]
```

```python
col_data.loc[
    (col_data['Private']==1) & 
    (col_data['Grad.Rate']>100)
, 'Grad.Rate'] = 100
```

```python
g = sns.FacetGrid(data = col_data, 
              hue = 'Private', palette = 'coolwarm', 
                  height = 6, aspect = 2, hue_order=[1, 0])
g.map(plt.hist, "Grad.Rate", bins = 20, alpha = 0.7, 
      edgecolor = 'black').add_legend()
plt.show()
```

## Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(col_data.drop('Private', axis = 1))
```

```python
kmeans.cluster_centers_
```

## Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(col_data['Private'], kmeans.labels_))
print(classification_report(col_data['Private'], kmeans.labels_))
```
