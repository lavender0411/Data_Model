# Advertisement
Dataset from kaggle.com

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ad_data = pd.read_csv('advertising.csv')
```

```python
ad_data.head()
```

```python
ad_data.info()
```

```python
ad_data.describe()
```

### Analysis
* Age 分佈

```python
sns.set_style('whitegrid')
sns.histplot(data = ad_data, x = 'Age', 
             bins = 30, edgecolor = 'black'
             )
plt.show()
```

* Area Income & Age 關係

```python
sns.jointplot(data = ad_data, x = 'Age', 
             y = 'Area Income'
             )
plt.show()
```

* Daily Time spent on site & Age 關係

```python
sns.jointplot(data = ad_data, x = 'Age', 
             y = 'Daily Time Spent on Site', 
             kind = 'kde', fill = True, color = 'red'
             )
plt.show()
```

* Daily Time Spent on Site & Daily Internet Usage 關係

```python
sns.jointplot(data = ad_data, y = 'Daily Internet Usage', 
             x = 'Daily Time Spent on Site', 
             )
plt.show()
```

* Clicked on Ad 與其他資料關係

```python
sns.pairplot(data = ad_data, hue = 'Clicked on Ad', diag_kind = 'hist')
plt.show()
```

### Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data.drop(['Ad Topic Line', 'City', 'Country', 'Timestamp', 'Clicked on Ad'], axis = 1)
y = ad_data['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter=300)
logmodel.fit(X_train, y_train)
```

### Predictions

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```
